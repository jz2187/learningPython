```python
def get_ali_lat_lng(address):
    lat_lng_list = list()
    import requests
    url="http://maps.googleapis.com/maps/api/geocode/json?address=%s" % (address)
    response = requests.get(url)
    if not response.status_code == 200:
        return None
    data = response.json()
    if not data['status'] == 'OK':
        return None
    #one_result = data['results'][0]
    for one_result in data['results']:
        latitude = one_result['geometry']['location']['lat']
        longtitude = one_result['geometry']['location']['lng']
        lat_lng_list.append((latitude, longtitude))
    return lat_lng_list
```
